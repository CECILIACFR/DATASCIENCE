#Detalhamento

#1.Busca pelos dados

O dataset escolhido neste trabalho é proveniente da plataforma **Kaggle**, sobre os **Livros mais vendidos (Top 50 bestselling) pela empresa Amazon de 2009-2019**, e que foram categorizados nos gêneros ficção e não-ficção. Para mais detalhes sobre este dataset, consulte: https://www.kaggle.com/datasets/sootersaalu/amazon-top-50-bestselling-books-2009-2019

O tipos de licença comun para este conjunto de dados é: Marca de domínio público.


Informação dos campos da **Tabela Livros** (dataset):

1.  **Name**: Nome do livro;
2.  **Author**: Autor do livro;
3.  **UserRating**: Avaliação/Classificação pelo usuário;
4.  **Reviews**: Número de avaliações escritas no Amazon;
5.  **Price**: Preço do livro;
6.  **Year**: Ano da realização da Avaliação pelo usuário;
7.  **Genre**: Gênero do livro.
